```python
from feature_engine.imputation import CategoricalImputer

# Specify the columns for imputation
columns_to_impute = ["Alley", "MasVnrType", "BsmtQual", "BsmtCond", "BsmtExposure", 
                     "BsmtFinType1", "BsmtFinType2", "Electrical", "FireplaceQu", 
                     "GarageType", "GarageFinish", "GarageQual", "GarageCond", 
                     "PoolQC", "Fence", "MiscFeature"]

# Create a subset DataFrame
df_subset = df[columns_to_impute].head()

# Initialize CategoricalImputer
imputer_cat = CategoricalImputer(imputation_method='missing', variables=columns_to_impute)

# Fit and transform the imputer on the subset
df[columns_to_impute] = imputer_cat.fit_transform(df[columns_to_impute])

# Check for missing values in the main DataFrame
print(df[columns_to_impute].isnull().sum())
```

```python
#categorical encoding
```

```python
from feature_engine.encoding import MeanEncoder

# Specify the columns for mean encoding
columns_to_encode = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
       'SaleType', 'SaleCondition']

# Create a copy of the original DataFrame
df_encoded = df.copy()

# Initialize MeanEncoder
encoder = MeanEncoder(variables=columns_to_encode)

# Fit and transform the encoder on the copy of the DataFrame
df_encoded[columns_to_encode] = encoder.fit_transform(df[columns_to_encode], df['SalePrice'])

```

```python
df_encoded.head()
```

```python
#feature scaling
```

```python
from sklearn.model_selection import train_test_split

x = df_encoded.drop('SalePrice', axis=1)
y = df['SalePrice']

x_reshaped = x.values.reshape((x.shape[0], x.shape[1], 1))

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn import preprocessing

std_scale = preprocessing.StandardScaler().fit(X_train)
X_train_std = std_scale.transform(X_train)
X_test_std = std_scale.transform(X_test)
```

# using deep learning model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU

model = Sequential([
    Dense(256, input_shape=(80,), activation='tanh'),
    Dense(128, activation='tanh'),
    Dense(128, activation='tanh'),
    Dense(64, activation='tanh'),
    Dense(64, activation='tanh'),
    Dense(64, activation='tanh'),  
    Dense(32, activation='tanh'),
    Dense(32, activation='tanh'),
    Dense(32, activation='tanh'),  
    Dense(16, activation='tanh'),  
    Dense(16, activation='tanh'),  
    Dense(1)
])
```

```python
print(model.summary())
```

```python
from tensorflow.keras.optimizers import Adam

# Reduce the learning rate of the Adam optimizer
optimizer = Adam(learning_rate=0.1)

# Compile the model with the new optimizer
model.compile(optimizer=optimizer, loss="mean_squared_error")
```

```python
#model.compile(optimizer='adam', loss="mean_squared_error")
```

```python
history = model.fit(X_train_std,y_train,epochs=150,batch_size=64,validation_split=0.2,verbose=1)
```

```python
plot_model(model,to_file='regression.png',show_shapes=True,show_layer_names=True)
```
